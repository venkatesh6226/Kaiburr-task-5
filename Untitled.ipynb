```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import os
import pandas as pd

# =======================
# CONFIG
# =======================
INPUT_PATH = '/Users/venkateshmakineni/Desktop/Projects/kaiburr/task-5/data/rows.csv'   # update if your file name differs
OUTPUT_PATH = "/Users/venkateshmakineni/Desktop/Projects/kaiburr/task-5/data/ccdb_balanced.csv"
os.makedirs("data", exist_ok=True)

# =======================
# LOAD DATA
# =======================
print("üì• Loading dataset...")
df = pd.read_csv(INPUT_PATH)
print("‚úÖ Loaded:", df.shape, "rows")

# Keep only needed columns
df = df[["Product", "Consumer complaint narrative"]]
df = df.dropna(subset=["Consumer complaint narrative"])

# =======================
# MAP PRODUCTS TO 4 CLASSES
# =======================
def map_category(p):
    s = str(p).lower()
    if "credit reporting" in s or "credit repair" in s:
        return "Credit reporting, repair, or other"
    if "debt collection" in s:
        return "Debt collection"
    if "consumer loan" in s:
        return "Consumer Loan"
    if "mortgage" in s:
        return "Mortgage"
    return None

df["canon_product"] = df["Product"].apply(map_category)
df = df[df["canon_product"].notnull()]
print("\nüìä Counts before balancing:\n", df["canon_product"].value_counts())

# =======================
# BALANCE CLASSES
# =======================
target_per_class = 10000  # adjust to 500, 1000, 2000 depending on available data
df_balanced = (
    df.groupby("canon_product", group_keys=False)
      .apply(lambda x: x.sample(min(len(x), target_per_class), random_state=42))
)

print("\nüìä Counts after balancing:\n", df_balanced["canon_product"].value_counts())

# Rename text column for training
df_balanced = df_balanced.rename(columns={"Consumer complaint narrative": "text"})

# Drop unnecessary columns
df_balanced = df_balanced[["text", "canon_product"]]

# =======================
# SAVE
# =======================
df_balanced.to_csv(OUTPUT_PATH, index=False)
print(f"\nüíæ Saved balanced dataset to {OUTPUT_PATH}")
print(f"üéØ Total complaints saved: {len(df_balanced)}")
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Train text classification models on the Consumer Complaint dataset
(4 categories: Credit Reporting, Debt Collection, Consumer Loan, Mortgage)
"""

import os
import re
import joblib
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
)

# =======================
# CONFIG
# =======================
DATA_PATH = "/Users/venkateshmakineni/Desktop/Projects/kaiburr/task-5/data/ccdb_balanced.csv"
ARTIFACTS_DIR = "artifacts"
os.makedirs(ARTIFACTS_DIR, exist_ok=True)

CANON_TO_ID = {
    "Credit reporting, repair, or other": 0,
    "Debt collection": 1,
    "Consumer Loan": 2,
    "Mortgage": 3,
}
ID_TO_CANON = {v: k for k, v in CANON_TO_ID.items()}


# =======================
# HELPERS
# =======================
def clean_text(s):
    """Basic cleaning for complaint text"""
    s = str(s).lower()
    s = re.sub(r"http\S+|www\.\S+", " ", s)
    s = re.sub(r"[^a-z\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s


def plot_confusion(y_true, y_pred, title):
    """Visualize confusion matrix"""
    cm = confusion_matrix(y_true, y_pred, labels=[0, 1, 2, 3])
    fig, ax = plt.subplots(figsize=(7, 6))
    im = ax.imshow(cm, interpolation="nearest", cmap="Blues")
    ax.set_title(title)
    ax.set_xlabel("Predicted")
    ax.set_ylabel("True")
    ax.set_xticks(range(4))
    ax.set_yticks(range(4))
    ax.set_xticklabels([ID_TO_CANON[i] for i in range(4)], rotation=45, ha="right")
    ax.set_yticklabels([ID_TO_CANON[i] for i in range(4)])
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, cm[i, j], ha="center", va="center", color="black")
    plt.tight_layout()
    plt.show()


# =======================
# MAIN
# =======================
if __name__ == "__main__":
    print("üì• Loading dataset...")
    df = pd.read_csv(DATA_PATH)
    print("‚úÖ Loaded:", len(df), "rows")

    # Map labels
    df["label"] = df["canon_product"].map(CANON_TO_ID)
    df["clean"] = df["text"].astype(str).apply(clean_text)
    df = df[df["clean"].str.split().str.len() >= 4]

    print("\nüìä Class distribution:")
    print(df["canon_product"].value_counts())

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        df["clean"], df["label"], test_size=0.2, random_state=42, stratify=df["label"]
    )

    # TF-IDF
    print("\n‚öôÔ∏è  Extracting features (TF-IDF)...")
    tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
    Xtr = tfidf.fit_transform(X_train)
    Xte = tfidf.transform(X_test)

    # =======================
    # TRAIN MODELS
    # =======================
    models = {
        "Linear SVM": LinearSVC(),
        "Logistic Regression": LogisticRegression(max_iter=1000),
        "Naive Bayes": MultinomialNB(),
    }

    results = {}

    print("\nüöÄ Training models...\n")
    for name, model in models.items():
        print(f"üîπ {name}")
        model.fit(Xtr, y_train)
        y_pred = model.predict(Xte)
        acc = accuracy_score(y_test, y_pred)
        print(f"Accuracy: {acc:.4f}")
        print(classification_report(
            y_test, y_pred,
            target_names=[ID_TO_CANON[i] for i in sorted(ID_TO_CANON)]
        ))
        results[name] = (model, y_pred, acc)
        print("-" * 60)

    # =======================
    # BEST MODEL
    # =======================
    best_name = max(results, key=lambda x: results[x][2])
    best_model, y_pred, acc = results[best_name]
    print(f"\n‚úÖ Best model: {best_name} (Accuracy = {acc:.4f})")

    # Plot confusion matrix
    plot_confusion(y_test, y_pred, best_name)

    # Save artifacts
    joblib.dump(best_model, f"{ARTIFACTS_DIR}/best_model.joblib")
    joblib.dump(tfidf, f"{ARTIFACTS_DIR}/tfidf.joblib")
    df.to_csv(f"{ARTIFACTS_DIR}/training_data_sample.csv", index=False)
    print(f"\nüíæ Artifacts saved in '{ARTIFACTS_DIR}/'")

    # =======================
    # QUICK DEMO
    # =======================
    def predict_text(texts):
        X = tfidf.transform([clean_text(t) for t in texts])
        preds = best_model.predict(X)
        return [ID_TO_CANON[p] for p in preds]

    demo = [
        "They keep reporting wrong information on my credit report.",
        "A debt collector is calling me multiple times a day.",
        "The mortgage company applied my payment to the wrong account.",
        "My consumer loan was denied due to false information.",
    ]

    print("\nüîç Demo predictions:")
    for t, p in zip(demo, predict_text(demo)):
        print(f"  - {p} :: {t}")
```
